```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1.

x_1 = np.arange(0,360)
x_2 = ( x_1 / 360 ) * 2 * np.pi

y = np.cos(x_2)

plt.plot(x_2,y)
plt.show
```

```python
plt.savefig("filename.png",dpi=300,format="png")
```
